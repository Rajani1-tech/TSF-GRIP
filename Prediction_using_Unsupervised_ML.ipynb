# K-Means Clustering with Scikit-Learn and the Iris Dataset

K-means clustering is a popular unsupervised machine learning technique used for grouping data points into clusters based on their similarity. In this project, we will apply K-means clustering to the famous Iris dataset to determine the optimal number of clusters and represent the results visually.

# Step 1: Importing all libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

# Loading the Dataset

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['Flower'] = iris.target
df.head()
```

# Applying K-Means Clustering

```python
km = KMeans(n_clusters=3)
yp = km.fit_predict(df)
yp
```

```python
df['cluster'] = yp
df.head(2)
```

```python
df.cluster.unique()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='skyblue')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='grey')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='pink')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
    plt.xlabel('K')

```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
